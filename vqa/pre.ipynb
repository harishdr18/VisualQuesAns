```python
import json

capt_train = json.load(open("D:/Project_phase_1/Preprocess/captions_train2014.json"))

coco = 'COCO_train2014_'

cap_dict = {}
for image_id, captions in capt_train.items():  # Iterate through items in capt_train
    if isinstance(captions, list):  # Ensure that captions is a list
        for caption in captions:  # Loop through each caption
            if caption is not None:  # Check if caption is not None
                if image_id not in cap_dict:
                    cap_dict[image_id] = [caption]
                else:
                    cap_dict[image_id].append(caption)
            else:
                print(f'No caption for image ID {image_id}')
```

```python
# Print the first 5 entries in cap_dict
for i, (image_id, captions) in enumerate(cap_dict.items()):
    if i < 5:  # Change this number to see more or fewer entries
        print(f'Image ID: {image_id}, Captions: {captions}')
    else:
        break
```

```python
with open('D:/Project_phase_1/Preprocess/cp_train_captions.json','w',encoding='utf-8') as file:
    file.write(json.dumps(cap_dict, indent=2, ensure_ascii=False))
```
